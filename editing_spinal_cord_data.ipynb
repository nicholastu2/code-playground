```python
import os
import numpy as np
from PIL import Image

def convert_tif_to_bmp(file):
    # Open the TIFF image
    with Image.open(file) as img:
        # Check the image mode and convert if necessary
        if img.mode == 'I;16' or img.mode == 'I;16B':
            # Convert the 16-bit image to an 8-bit image by scaling the pixel values
            img = np.array(img, dtype=np.uint16)
            img = (img / 256).astype(np.uint8)
            img = Image.fromarray(img)
        bmp_file = file.replace('.tif', '.bmp')
        img.save(bmp_file, format='BMP')
    return bmp_file

path1 = r"S:\Phys\FIV925 XSection\Datasets\SpinalAtlas\Dodd and Fiederling\Cropped\2_Inverted_Slides\I2"
path2 = r"S:\Phys\FIV925 XSection\Datasets\SpinalAtlas\Dodd and Fiederling\Cropped\8_Inverted_Slides\I8"

list_of_paths = [path1, path2]

for path in list_of_paths:
    # List all files in the directory
    for file in os.listdir(path):
        # Check if the file is a TIFF file
        if file.endswith('.tif'):
            full_path = os.path.join(path, file)
            convert_tif_to_bmp(full_path)
            print(f"Converted {full_path} to BMP")
```

## checking whether the image is multi-dimensional or not

```python
from PIL import Image

def is_multichannel(file):
    # Open the image
    with Image.open(file) as img:
        # Get the image mode
        mode = img.mode
        # Determine if the image is multichannel
        if mode in ['RGB', 'RGBA', 'CMYK']:
            return True
        else:
            return False

# Example usage
bmp_file = r"S:\Phys\FIV925 XSection\Datasets\SpinalAtlas\Dodd and Fiederling\Cropped\8_Inverted_Slides\I8\C3-Slide1-6_Region0007_Channel555 nm,475 nm,395 nm_Seq0051.tif"
if is_multichannel(bmp_file):
    print(f"The image is multichannel.")
else:
    print(f"The image is single-channel.")
```

## Drawing polygons on blank image

```python
import numpy as np
from PIL import Image, ImageDraw

# Function to read and parse the coordinates from the text file
def read_coordinates(file_path):
    with open(file_path) as file:
        data = file.readlines()
    
    polygons = []
    for line in data:
        # Split the line into coordinates and convert to float
        coords = list(map(float, line.strip().split()[1:]))
        # Reshape the coordinates into pairs (x, y)
        polygon = np.array(coords).reshape(-1, 2)
        polygons.append(polygon)
    return polygons

# Function to draw normalized polygons on the image
def draw_polygon_normalized(draw, polygon, image_width, image_height):
    scaled_polygon = [(x * image_width, y * image_height) for x, y in polygon]
    draw.polygon(scaled_polygon, outline="red", width=2)

# Main function to create a new image and draw polygons
def create_image_with_polygons(coordinates_file, output_image_path, image_width=800, image_height=600):
    polygons = read_coordinates(coordinates_file)

    # Create a new blank image (white background)
    new_image = Image.new("RGB", (image_width, image_height), "white")

    # Create a drawing context
    draw = ImageDraw.Draw(new_image)

    # Draw all polygons with normalization
    for polygon in polygons:
        draw_polygon_normalized(draw, polygon, image_width, image_height)

    # Save the new image
    new_image.save(output_image_path)
    print(f"Image saved to {output_image_path}")

# Example usage
coordinates_file = r"S:\Phys\FIV925 XSection\Datasets\SpinalAtlas\Dodd and Fiederling\Cropped\8_Inverted_Slides\I8\C2-Slide1-6_Region0007_Channel555 nm,475 nm,395 nm_Seq0051.txt"  # Replace with your coordinates file path
output_image_path = r"S:\Phys\FIV925 XSection\Datasets\SpinalAtlas\Dodd and Fiederling\Cropped\8_Inverted_Slides\I8\new_image_with_polygons.bmp"
create_image_with_polygons(coordinates_file, output_image_path)
```
