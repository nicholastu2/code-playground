```python
import numpy as np
from scipy.spatial import procrustes
from scipy.interpolate import interp1d
from PIL import Image, ImageDraw

# Function to read and parse the coordinates from the text file
def read_coordinates(file_path):
    with open(file_path) as file:
        data = file.readlines()
    
    polygons = []
    for line in data:
        # Split the line into coordinates and convert to float
        coords = list(map(float, line.strip().split()[1:]))
        # Reshape the coordinates into pairs (x, y)
        polygon = np.array(coords).reshape(-1, 2)
        polygons.append(polygon)
    return polygons

# Function to resample a polygon to have a specific number of points
def resample_polygon(polygon, num_points):
    t = np.linspace(0, 1, len(polygon))
    t_resampled = np.linspace(0, 1, num_points)
    interp_x = interp1d(t, polygon[:, 0], kind='linear')
    interp_y = interp1d(t, polygon[:, 1], kind='linear')
    resampled_polygon = np.vstack((interp_x(t_resampled), interp_y(t_resampled))).T
    return resampled_polygon

# Function to perform Procrustes analysis
def procrustes_analysis(src, dst):
    mtx1, mtx2, disparity = procrustes(src, dst)
    return mtx2

# Function to draw polygons on an image
def draw_polygon(draw, polygon):
    draw.polygon(polygon, outline="red", width=2)

# Main function to morph polygons to fit a target shape
def morph_polygons_to_shape(coordinates_file, target_shape, output_image_path, image_width=800, image_height=600):
    polygons = read_coordinates(coordinates_file)

    # Create a new blank image (white background)
    new_image = Image.new("RGB", (image_width, image_height), "white")
    draw = ImageDraw.Draw(new_image)

    for polygon in polygons:
        # Resample the polygon to match the number of points in the target shape
        resampled_polygon = resample_polygon(polygon, len(target_shape))
        
        # Perform Procrustes analysis to fit polygon to the target shape
        transformed_polygon = procrustes_analysis(resampled_polygon, target_shape)
        
        # Scale and draw the transformed polygon
        scaled_polygon = [(x * image_width, y * image_height) for x, y in transformed_polygon]
        draw_polygon(draw, scaled_polygon)

    # Save the new image
    new_image.save(output_image_path)
    print(f"Image saved to {output_image_path}")

# Example usage
coordinates_file = r"S:\Phys\FIV925 XSection\Datasets\SpinalAtlas\Dodd and Fiederling\Cropped\8_Inverted_Slides\I8\C2-Slide1-6_Region0007_Channel555 nm,475 nm,395 nm_Seq0051.txt"  # Replace with your coordinates file path
output_image_path = r"S:\Phys\FIV925 XSection\Datasets\SpinalAtlas\Dodd and Fiederling\Cropped\8_Inverted_Slides\I8\morphed_polygons_image.png"  # Replace with your output image path  # Replace with your output image path

# Define the target shape (e.g., a square)
target_shape = np.array([[200, 200], [500, 200], [500, 500], [200, 500]])

morph_polygons_to_shape(coordinates_file, target_shape, output_image_path)
```

```python
import cv2
import numpy as np

def extract_red_polygons(image_path):
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading image: {image_path}")
        return []

    # Convert the image to HSV color space
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define the red color range
    lower_red1 = np.array([0, 70, 50])
    upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([170, 70, 50])
    upper_red2 = np.array([180, 255, 255])

    # Create masks for red color
    mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
    mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
    mask = cv2.bitwise_or(mask1, mask2)

    # Find contours in the mask
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Filter out small contours (noise)
    min_contour_area = 100  # Adjust this value based on your image
    contours = [cnt for cnt in contours if cv2.contourArea(cnt) > min_contour_area]

    # Get the coordinates of the polygons
    polygons = [cnt.reshape(-1, 2).tolist() for cnt in contours]

    return polygons

def draw_polygons(polygons, canvas_size, output_path):
    # Create a blank canvas
    canvas = np.ones((canvas_size[1], canvas_size[0], 3), dtype=np.uint8) * 255

    # Draw each polygon on the canvas
    for polygon in polygons:
        pts = np.array(polygon, np.int32)
        pts = pts.reshape((-1, 1, 2))
        cv2.polylines(canvas, [pts], isClosed=True, color=(0, 0, 255), thickness=2)

    # Save the resulting image
    cv2.imwrite(output_path, canvas)
    print(f"Polygons drawn and saved to {output_path}")

# Example usage
image_path = r"S:\Phys\FIV925 XSection\Datasets\SpinalAtlas\Dodd and Fiederling\Cropped\8_Inverted_Slides\I8\C2-Slide1-6_Region0007_Channel555 nm,475 nm,395 nm_Seq0051_adjusted4.png"  # Replace with your image path
output_path = r"S:\Phys\FIV925 XSection\Datasets\SpinalAtlas\Dodd and Fiederling\Cropped\8_Inverted_Slides\I8\output_image.png"  # Replace with your output image path
canvas_size = (800, 600)  # Replace with the desired canvas size (width, height)

# Extract polygons
polygons = extract_red_polygons(image_path)

# Draw polygons on a blank canvas
if polygons:
    draw_polygons(polygons, canvas_size, output_path)
else:
    print("No polygons found.")
```
